```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_excel (r'./1.xlsx')
print (df)
```

```python
threedee = plt.figure().gca(projection='3d')
threedee.scatter(df['DaysUSAScreened'], df['Durationmins'], df['Globalreceipts($m)'])
threedee.set_xlabel('DaysUSAScreened')
threedee.set_ylabel('Durationmins')
threedee.set_zlabel('Globalreceipts($m)')
plt.tight_layout()
plt.show()
```

```python
threedee = plt.figure().gca(projection='3d')
threedee.scatter(df['Production$m'], df['TotalTheaters'], df['Globalreceipts($m)'])
threedee.set_xlabel('Production$m')
threedee.set_ylabel('TotalTheaters')
threedee.set_zlabel('Globalreceipts($m)')
plt.tight_layout()
plt.show()
```
